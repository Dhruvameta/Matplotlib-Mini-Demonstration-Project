```python
import matplotlib.pyplot as plt
```

**Chart**

```python
#data 

x = [1,2,3,4,5,]
y = [10,20,25,35,45]

plt.plot(x,y) #line chart
plt.show() # To display chart
```

**Customized chart**

```python
plt.plot(x,y, color='r', marker='o',linestyle='dashed', linewidth=2, markersize=12)
plt.show()
```

```python
plt.figure(figsize=(4,3)) # changing size of the chart
plt.plot(x,y, color='r', marker='o',linestyle='dashed', linewidth=2, markersize=12)
plt.show()
```

**ADDING A HEADING TO CHART**

```python
plt.figure(figsize=(6,4)) # changing size of the chart
plt.plot(x,y, color='g', marker='o',linestyle='dashed', linewidth=2, markersize=12)

plt.title("Growth")
plt.xlabel("consistency")
plt.ylabel("Hardwork")

plt.show()
```

**Advance line charts - (multiple lines & legends)**

```python
x = [1,2,3,4,5,]
y1 = [10,20,25,35,45]
y2 = [20,30,45,50,55]

plt.plot(x,y1,color='r', label="sales 2024")
plt.plot(x,y2, color='g', label="sales 2025")

plt.title("Yealy sales")
plt.xlabel("months")
plt.ylabel("sales")

plt.legend()
plt.show()
```

**BAR CHART**

```python
x = ['A','B','C','D','E',]
y = [10,20,25,35,45]

plt.bar(x,y)
plt.title('Bar chart')
plt.xlabel("months")
plt.ylabel("sales")

plt.show
```

**HISTOGRAM**

```python
# used for distribution analysis

import random
data = [random.randint(1,30)for _ in range(500)] 

plt.hist(data, bins=20)
plt.title("HISTOGRAM")
plt.show()
```

**PIE CHART**

```python
# used to show part to whole relationship/data

categories = ['A','B','C','D','E',]
sales = [10,20,25,35,45]

plt.pie(sales, labels=categories , autopct= '%1.1f%%')
plt.title("PIE CHART EXAMPLE")
plt.show()
```

```python

categories = ['A','B','C','D','E',]
sales = [36,40,64,50,45]

plt.pie(sales, labels=categories , autopct= '%1.1f%%', startangle=90)
plt.title("PIE CHART EXAMPLE")
plt.show()
```

**SCATTER PLOT**

```python
# scatter plot
# used to show part to whole relationship/data

y1 = ['A','B','C','D','E',]
y2 = [10,20,25,35,45]

plt.scatter(y1,y2)
plt.title("SCATTER PLOT")
plt.show()

```

**SUBPLOTS**

```python
# used to show multiple charts in one figure 

# data-1 (bar chart)
categories = ['Mon','Tue','Wed','Thu','Fri']
sales = [36,40,64,50,45]

# data-2 (scatter chart)
y1 = ['A','B','C','D','E',]
y2 = [10,20,25,35,45]

plt.Figure(figsize=(10,4))

plt.subplot(1,2,2) # row, column, position
plt.bar(categories, sales)
plt.title("Daily sales")
plt.xlabel("week days")
plt.ylabel("sales")


plt.subplot(1,2,1) # row, column, position
plt.scatter(y1,y2)
plt.title("USERS")
plt.xlabel("USER1")
plt.ylabel("USER2")

plt.show()


```
